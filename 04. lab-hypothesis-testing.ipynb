In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#h0: HP_stats_dragon <= HP_stats_Grass
#h1: hp_stats_dragon > hp_stats_grass
df_dragon = df[df["Type 1"] == "Dragon"]["HP"]
df_grass = df[df["Type 1"] == "Grass"]["HP"]

#alpha = 0.05

st.ttest_ind(df_dragon,df_grass, alternative="greater")
#p-value is lower than alpha so we can reject the null hypothesis.
```

```python
df.columns
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
#h0: leg_pok_stats == non-leg_pok_stats
#h1: leg_pok_stats != non-leg_pok_stats

#aplha = 0.05
leg_df = df[df["Legendary"] == True][["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]
non_leg_df = df[df["Legendary"] == False][["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]

#anova test:
st.f_oneway(leg_df, non_leg_df)

#for each stat, we we very low p-values, so we can reject the null hypothesis
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
#Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
import scipy.spatial.distance as dist

def calculate_distance(latitude, longitude):
    
    #latitude = df["latitude"]
    #longitude = df["longitude"]

    school_distance = dist.euclidean((longitude, latitude), (-118,37))
    hospital_distance = dist.euclidean((longitude, latitude), (-122, 34))


    #print(school_distance)
    #print(hospital_distance)

    dista = min(school_distance, hospital_distance)


    if dista < 0.5:
        return True
    else:
        return False
```

```python
#calculate the euclidean distance from each house to the school and to the hospital:
df['close_far'] = df.apply(lambda row: calculate_distance(row['latitude'], row['longitude']), axis=1)
df
```

```python
# Divide your dataset into houses close and far from either a hospital or school.
close_df = df[df["close_far"] == True]["median_house_value"].reset_index(drop=True).to_frame()
far_df = df[df["close_far"] == False]["median_house_value"].reset_index(drop=True).to_frame()
print(close_df)
print(far_df)
close_df.head()
```

```python
#Choose the propper test and, with 5% significance, comment your findings.

#H0: close_value ==far_value
#H1: close_value != far_value

#alpha = 0.05

st.ttest_ind(close_df["median_house_value"],far_df["median_house_value"])
```
